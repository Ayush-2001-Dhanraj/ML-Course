```python
heart_data = pd.read_csv("heart-disease.csv")
heart_data.head()
```

```python
heart_data.plot(subplots = True, figsize = (10, 30));
```

```python
heart_data.plot.hist(subplots = True, figsize = (5, 30));
```

```python
heart_data.head()
```

```python
over_50 = heart_data[heart_data["age"] > 50]
over_50.head()
```

```python
# using pyplot method
over_50.plot(kind = "scatter",
             figsize = (15, 5),
             x = "age",
             y = "chol",
             c = "target");
```

```python
# using matplotlib OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (15, 5))

over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target",
             ax = ax);

# ax.set_xlim([45, 100])
```

```python
# using matplotlib OO from scratch

fig, (ax1, ax2) = plt.subplots(figsize = (15, 5), 
                                 ncols = 2, 
                                 nrows = 1)

scatter_ax1 = ax1.scatter(x = over_50["age"],
                          y = over_50["chol"],
                          c = over_50["target"])

scatter_ax2 = ax2.scatter(x = over_50["age"],
                          y = over_50["thalach"],
                          c = over_50["target"])

ax1.set(title = "Cholestrol",
         xlabel = "Age",
         ylabel = "Cholestrol")

ax2.set(title = "Heart Rate",
        xlabel = "Age",
        ylabel = "Heart Rate")

ax1.legend(*scatter_ax1.legend_elements(), title = "target")

ax2.legend(*scatter_ax2.legend_elements(), title = "target")

ax1.axhline(over_50["chol"].mean(), linestyle = "--");

ax2.axhline(over_50["thalach"].mean(), linestyle = ":");

fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

```python
# using matplotlib OO from scratch

plt.style.use('seaborn-v0_8-whitegrid')

fig, (ax1, ax2) = plt.subplots(figsize = (15, 5), 
                                 ncols = 2, 
                                 nrows = 1)

scatter_ax1 = ax1.scatter(x = over_50["age"],
                          y = over_50["chol"],
                          c = over_50["target"],
                          cmap = "inferno")

scatter_ax2 = ax2.scatter(x = over_50["age"],
                          y = over_50["thalach"],
                          c = over_50["target"],
                          cmap = "inferno")

ax1.set(title = "Cholestrol",
         xlabel = "Age",
         ylabel = "Cholestrol")

ax2.set(title = "Heart Rate",
        xlabel = "Age",
        ylabel = "Heart Rate")

ax1.legend(*scatter_ax1.legend_elements(), title = "target")

ax2.legend(*scatter_ax2.legend_elements(), title = "target")

ax1.axhline(over_50["chol"].mean(), linestyle = "--");

ax2.axhline(over_50["thalach"].mean(), linestyle = ":");

ax1.set_xlim([50, 80])
ax2.set_xlim([50, 80])

ax2.set_ylim([60, 200])

fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
