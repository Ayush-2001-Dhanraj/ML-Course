## Matplotlib workflow sample

```python
## imports

## data prepataion
x = [2, 4, 6, 8]
y = [2, 4, 16, 64]

## setup plot
fig, ax = plt.subplots(figsize = (5, 5))

## plot data
ax.plot(x, y)

## cutomize plot
ax.set(title = "Simple plot", xlabel = "x - axis", ylabel = "y - axis")

## save & show plot
fig.savefig("simple_plot.png")
```

```python
butter = {"Almond Butter": 10, "Peanut Butter": 8, "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(butter.keys(), butter.values())

ax.set(title = "Ayush butter show", ylabel = "Cost (RS.)");
```

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x);
```

## Subplots option 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

ax1.plot(x, x / 2)
ax2.scatter(np.random.randint(10, size = 10), np.random.randint(10, size = 10))
ax3.bar(butter.keys(), butter.values())
ax4.hist(x);
```

## Subplots option 2

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))
ax[0, 0].plot(x, x ** 2)
ax[0, 1].scatter(np.random.random(100), np.random.random(100))
ax[1, 0].barh(list(butter.keys()), list(butter.values()))
ax[1, 1].hist(np.random.randn(1000));
```

## Visualizing from dataframes

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range("1/19/2024", periods = 1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,]', "", regex = True).astype(float).astype(int)
car_sales
```

```python
car_sales["Total"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("2/19/2024", periods = len(car_sales))
car_sales
```

```python
car_sales.plot("Sale Date", "Total");
```

```python
car_sales.plot("Odometer (KM)", "Price", kind = "scatter");
```
